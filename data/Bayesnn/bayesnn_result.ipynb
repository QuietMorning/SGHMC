```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
log = open('./demo/log.txt','r')
log_sgd = open('./demo/log_sgd.txt','r')
log_sgld = open('./demo/log_sgld.txt','r')
```

```python
# Results for three algorithms

result = log.read()
result_sgd = log_sgd.read()
result_sgld = log_sgld.read()
```

```python
# Test errors for three algorithms

test_errs = np.array([float(item[9:]) for item in result.split() if item.startswith('test-err')])
test_errs_sgd = np.array([float(item[9:]) for item in result_sgd.split() if item.startswith('test-err')])
test_errs_sgld = np.array([float(item[9:]) for item in result_sgld.split() if item.startswith('test-err')])
```

```python
plt.plot(test_errs[5:],c='r',label="SGHMC")
plt.plot(test_errs_sgd[5:],c='b',label="SGD")
plt.plot(test_errs_sgld[5:],c='g',label="SGLD")
plt.xlabel("steps")
plt.ylabel("test error")
plt.title("bayesian neural network test errors for three algorithms")
plt.legend()
pass
```
